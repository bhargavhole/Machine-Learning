```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

Another way to load the dataset is using seaborn  

```python
df = sns.load_dataset('iris')

# here we are getting better maintained dataset so for the independent features we will use sns and dependent features (sklearn dataset)
```

INDEPENDENT AND DEPENDENT FEATURES:

```python
X = df.iloc[:,:-1]      # iloc[all ROWS, all columns -except last column] ...... iloc[rows, columns]
y = iris.target         # Here we used the sklearn dataset 
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=42)
```

```python
# post prunning - first we will make tree and then prun it.
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train, y_train)
```

In the above figure we can see - at level 2 -> value = [0,35,1] and value = [0,5,37] we can see huge difference b/w values so we can stop the tree at that level, so we will now put max_depth parameter in "treemodel = DecisionTreeClassifier(max_depth=2)"

```python
from sklearn import tree
plt.Figure(figsize=(30, 25))
tree.plot_tree(treemodel, filled=True)
```

```python
y_pred = treemodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred, y_test)
score
```

```python
print(classification_report(y_pred, y_test))
```

From here Pre prunning part begins

```python
# Pre Prunning

parameter = {
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
cv = GridSearchCV(treemodel, param_grid=parameter, cv=5, scoring='accuracy')
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
score = accuracy_score(y_pred, y_test)
score
```

```python
print(classification_report(y_pred, y_test))
```
